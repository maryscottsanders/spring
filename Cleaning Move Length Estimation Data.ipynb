```python
import pandas as pd
import numpy as np
```

```python
#Read csv into a pandas dataframe.
csv = 'move_length.csv'
df = pd.read_csv(csv, index_col = 'order_id')
df.head()
```

```python
df.info()
```

```python
#Removing columns from the 'calculator.'
#These columns come from a new pre-move survey we only give to our phone customers that asks things like how many flights of stairs or large appliances there are in their home.
#They all have less than 500 non-null observations, and there is no sensical way to replace these missing values, so I will remove these columns. (Maybe in the future, this analysis can be re-visited and incorporate these columns.)
df = df.drop(['refrigerators','some_calc','move_type','calculator_count_of_hops','calculator_count_of_rooms','calculator_count_of_boxes','calculator_count_of_items','calculator_from_has_basement','calculator_from_count_of_garage_stalls','calculator_to_has_elevator','calculator_from_has_elevator','calculator_hours'], axis='columns')
df.info()
```

```python
#Drop rows where move_length_hours is null. 
#This is the column from which the independant variable is calculated. 
#It would not make sense to replace missing values and there is only a very small number of these.
df = df[df['move_length_hours'].notnull()]
df = df[df['move_length_hours'] > 0]
df.info()
```

```python
df.property_type_name = df.property_type_name.astype('category')
df.info()
```

```python
#Add a column for whether or not move exceeded 4 hours in length
df['over'] = np.where(df.move_length_hours > 4,1,0)
df.describe()
```

```python
#Transform reservation_start and booked_at to datetimes
df['reservation_start'] = pd.to_datetime(df['reservation_start'])
df['booked_at'] = pd.to_datetime(df['booked_at'])
df.info()
```

```python
#Add a column for time of day
df['hour_of_day'] = df['reservation_start'].dt.hour

#Add a column for morning
df['morning'] = np.where(df['hour_of_day']<=12,1,0)

#Add a column for day of week (The day of the week with Monday=0, Sunday=6)
df['day_of_week'] = df['reservation_start'].dt.dayofweek

#Add a column for weekend
df['weekend'] = np.where(df['day_of_week']>=4 ,1,0)

#Transform day of week to category
df['day_of_week'] = df['reservation_start'].dt.dayofweek.astype('category')

df.info()
```
